## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X_train_reduzido, X_val, y_train_reduzido, y_val = train_test_split(X_train_norm, y_train, test_size=0.2, random_state=42)
```

#### Descobrindo a melhor quantidade de vizinhos prximos "k" para o modelo.

```python
k_neighbors = [x for x in range(2,30)]
best_mae = np.Inf
best_k = 0


for k in k_neighbors:
  knn = KNeighborsRegressor(k)
  knn.fit(X_train_reduzido, y_train_reduzido)

  y_pred_val = knn.predict(X_val)
  mae = MAE(y_val, y_pred_val)
  if mae < best_mae:
    print(f'valor de k: {k} tem MAE de {mae}')
    best_mae = mae
    best_k = k
```

```python
best_k
```

```python
knn = KNeighborsRegressor(best_k)
knn.fit(X_train_norm, y_train)
```

```python
y_pred_train = knn.predict(X_train_norm)
y_pred_test = knn.predict(X_test_norm)
```

```python
print(f"RMSE em Treinamento {np.sqrt(MSE(y_train, y_pred_train)) }")
print(f"RMSE em Teste {np.sqrt(MSE(y_test, y_pred_test)) }")

print(f"MAE em Treinamento {MAE(y_train, y_pred_train)}")
print(f"MAE em Teste {MAE(y_test, y_pred_test)}")
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
criterios = ['squared_error', 'absolute_error']
max_dephts = [1, 5, 10, 20, 50, 100]

best_mae = np.Inf
best_params = None


for criterio in criterios:
  for max_depth in max_dephts:
    dt = DecisionTreeRegressor(criterion = criterio, max_depth = max_depth)
    dt.fit(X_train_reduzido, y_train_reduzido)
    y_pred_val = dt.predict(X_val)
    mae = MAE(y_val, y_pred_val)

    if mae < best_mae:
      best_mae = mae
      best_params = (criterio, max_depth)
      print(f'valor de criterio: {criterio} e max_depth: {max_depth} tem MAE de {mae}')
```

```python
best_params
```

```python
dt = DecisionTreeRegressor(criterion = best_params[0], max_depth = best_params[1])
dt.fit(X_train_norm, y_train)
```

```python
y_pred_train = dt.predict(X_train_norm)
y_pred_test = dt.predict(X_test_norm)
```

```python
print(f"RMSE em Treinamento {np.sqrt(MSE(y_train, y_pred_train)) }")
print(f"RMSE em Teste {np.sqrt(MSE(y_test, y_pred_test)) }")

print(f"MAE em Treinamento {MAE(y_train, y_pred_train)}")
print(f"MAE em Teste {MAE(y_test, y_pred_test)}")
```
